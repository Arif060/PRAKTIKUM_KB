```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
import tensorflow as tf
from tensorflow.keras import layers, models
import matplotlib.pyplot as plt

# 1. Memuat dataset
anime_data = pd.read_csv('anime.csv')

# 2. Menampilkan beberapa baris pertama untuk memastikan data dimuat dengan benar
print(anime_data.head())

# 3. Memeriksa apakah ada nilai yang hilang
print(anime_data.isnull().sum())

# 4. Menghapus baris dengan nilai yang hilang (jika ada)
anime_data = anime_data.dropna()

# 5. Mengonversi data kategorikal (tipe) menjadi numerik menggunakan LabelEncoder
label_encoder = LabelEncoder()
anime_data['type'] = label_encoder.fit_transform(anime_data['type'])

# 6. Memisahkan kolom genre menjadi beberapa kolom biner untuk setiap genre
genres = anime_data['genre'].str.get_dummies(sep=', ')

# 7. Memilih fitur dan target
features = pd.concat([anime_data[['type', 'members']], genres], axis=1)

# 8. Menstandarkan kolom 'members'
scaler = StandardScaler()
features['members'] = scaler.fit_transform(features[['members']])

# 9. Mengonversi target menjadi kategori (0: rendah, 1: sedang, 2: tinggi)
bins = [0, 5, 7.5, 10]  # Batas untuk rendah, sedang, tinggi
labels = [0, 1, 2]  # Label untuk kategori
anime_data['rating_category'] = pd.cut(anime_data['rating'], bins=bins, labels=labels)

# Memperbarui target untuk menjadi kategorikal
target = anime_data['rating_category'].astype(int)  # Pastikan target adalah integer

# 10. Memisahkan dataset menjadi training dan validation sets
X_train, X_val, y_train, y_val = train_test_split(features, target, test_size=0.2, random_state=42)

# 11. Membangun model deep learning
model = models.Sequential([
    layers.Dense(64, activation='relu', input_shape=(X_train.shape[1],)),
    layers.Dense(32, activation='relu'),
    layers.Dense(3, activation='softmax')  # Output untuk 3 kategori
])

# 12. Mengompilasi model dengan loss sparse_categorical_crossentropy dan metrik akurasi
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# 13. Melatih model
history = model.fit(X_train, y_train, epochs=50, validation_data=(X_val, y_val), batch_size=32)

# 14. Mengevaluasi model
train_loss, train_accuracy = model.evaluate(X_train, y_train)
val_loss, val_accuracy = model.evaluate(X_val, y_val)

# 15. Menampilkan hasil akurasi
print(f"Akurasi Pelatihan: {train_accuracy * 100:.2f}%, Akurasi Validasi: {val_accuracy * 100:.2f}%")

# 16. Menampilkan diagram loss dan akurasi
plt.figure(figsize=(12, 5))

# Plotting Loss
plt.subplot(1, 2, 1)
plt.plot(history.history['loss'], label='Loss Pelatihan')
plt.plot(history.history['val_loss'], label='Loss Validasi')
plt.title('Loss Pelatihan dan Validasi')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

# Plotting Akurasi
plt.subplot(1, 2, 2)
plt.plot(history.history['accuracy'], label='Akurasi Pelatihan')
plt.plot(history.history['val_accuracy'], label='Akurasi Validasi')
plt.title('Akurasi Pelatihan dan Validasi')
plt.xlabel('Epoch')
plt.ylabel('Akurasi')
plt.legend()

plt.show()
```
